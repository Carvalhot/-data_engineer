# Visualização de dados
## Criando legendas e anotações
___

```python
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns


%matplotlib inline
rcParams['figure.figsize'] = 16, 8
sns.set_style('whitegrid')

caminho = 'C:/Users/luizc/Desktop/ENGENHARIA DE DADOS/LINKEDIN - PYTHON/MANIPULAÇÃO/dados/mtcars.csv'
carros = pd.read_csv(caminho)
carros.columns = ['nomes','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
mpg = carros['mpg']
```

### Colocando títulos nos eixos

```python
x = range(1, 10)
y = [1, 2, 3, 4, 0.5, 4, 3, 2, 1]
plt.bar(x, y)

plt.xlabel('legenda no eixo X')
plt.ylabel('legenda no eixo Y')
plt.show()
```

### Colocalndo legenda no gráfico

```python
import matplotlib.patches as mpatches

plt.plot(mpg)
plt.title('Consumo de gasolina')
plt.xlabel('nomes dos carros')
plt.ylabel('millhas/galão')
#passando o nome dos carros (ticks) para o gráfico
plt.xticks(mpg.index, carros.nomes, rotation=45)

#Configurando legenda de cor
legenda = mpatches.Patch(label='mpg')
plt.legend(handles=[legenda])

plt.show()
```

### Fazendo anotações no gráfico
